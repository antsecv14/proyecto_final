```python
#Se importan todas las librerias a usar dentro de la 
import pandas as pd
from math import pi
from bokeh.core.properties import value
from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure
from bokeh.transform import cumsum     
from bokeh.palettes import *
#utolozamos el output notebook para que la grafica no se abra en otra pestaña
output_notebook()
#Mediante pandas leemos nuestro csv
datos = pd.read_csv('data/bdd-transportes-2012.csv', sep=';', encoding="ISO-8859-1")
#Declaramos una lista de colores
colors01 = ['#FF0000', '#B22222', '#8B0000', '#FF1493', '#C71585', '#DB7093', '#FF4500', '#FF8C00', '#FFA500', '#6A5ACD', 
            '#483D8B', '#7B68EE', '#ADFF2F', '#7FFF00', '#7CFC00', '#00008B', '#191970', '#813C3C', '#FFD8EE', '#FFE7D8', 
            '#D8DFFF', '#D8FAFF', '#D8FFE2', '#F3FFD8', '#FFE7D8', '#6B6464', '#813C3C']
group  = datos.groupby('DESPROVINCIA',as_index = False).mean()
#Declaramos dos variables provincias y f_provincias en la cual identificaremos las provincias dentro del atrivuto provincias y
#contamos las veces que se repite cada provincia
provincias, f_provincias = np.unique(datos['DESPROVINCIA'], return_counts=1)
#Creamos la grafica
figure01 = figure(x_range=provincias, plot_width=2000, title = "NUMERO DE TRANSPORTES POR PROVINCIA")
#con title.text_font_size para poder determinar el tama`o del titulo
figure01.title.text_font_size = "18pt"
#con xaxis.major_label_orientation determinamos la rotacion de los atributos 
figure01.xaxis.major_label_orientation = 1.1
#Finalmente determinamos los valores de los ejes de la grafica los colores a usar y la presentamos
figure01.vbar(x=provincias , top=f_provincias, width=0.9, line_color=colors01[16], line_width=0.5, fill_color=colors01[0:len(provincias)])
show(figure01)




```

```python

#llevamos a cabo el mismo proceso de la anterior grafica
combustible, f_combustible = np.unique(datos['DESCOMBUSTIBLE'], return_counts=1)
figure02 = figure(y_range=combustible, plot_width=2000, title = "NUMERO DE TRANSPORTES POR COMBUSTIBLE")
figure02.title.text_font_size = "18pt"
figure02.xaxis.major_label_orientation = 1.1
#la unica diferencia es que cambiamos nuestro eje x por el y y en ves de usar top utilizamos right 
figure02.hbar(y=combustible , right=f_combustible, height=0.9, line_color=colors01[16], line_width=0.5, fill_color=colors01[0:len(combustible)])
show(figure02)
```

```python

descombustible, f_descombustible = np.unique(datos['DESCOMBUSTIBLE'], return_counts=1)
figure03 = figure(x_range = descombustible, plot_width = 800, title = "NUMERO DE TRANSPORTES POR COMBUSTIBLE")
figure03.title.text_font_size = "18pt"
figure03.xaxis.major_label_orientation = 1.1
figure03.vbar(x= descombustible , top=f_descombustible, width=0.5, line_color=colors01[16], line_width=0.5, fill_color=colors01[0:len(descombustible)])
show(figure03)
```

```python


#Creamos un contador de cuantos modelos por ano existen 
contador_modelos = datos.groupby(['DESMODELO'])['DESMODELO'].count() 
#data sera la informacion que se le enviara la figura es decir los anos que existen y la cantidad de transportes que existen
data = pd.Series(contador_modelos).reset_index(name='value').rename(columns={'index':'count'})
#obtenemos el angulo que cubrira el area de cada ano
data['angle'] = data['value']/data['value'].sum() * 2*pi
#declaramos los colores
data['color'] = Category20c[len(contador_modelos)]
#determinamos el tamano de la grafica su titulo
p = figure(plot_height=350, title="TRANSPORTES GRAFICADOS POR MODELO", toolbar_location=None,
           tools="hover", tooltips="@country: @value", x_range=(-0.5, 1.0))
#Se determinan los parametros de la figura 
p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend='DESMODELO', source=data)


show(p)
```
